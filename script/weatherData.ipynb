```python
# Documentation: https://dev.meteostat.net/python/
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily, Hourly, Stations
```

```python
# Set time period
start = datetime(2020, 1, 2)
end = datetime(2023, 9, 30)
# Create Point for some place
# somePlace = Point(52.5, 13.5)
# get data from nearest weather station
# - Bottrop-Welheim #https://www.lanuv.nrw.de/luqs/messorte/steckbrief.php?ort=BOTT; 51.525955, 6.976880
# - Duisburg-Walsum #https://www.lanuv.nrw.de/luqs/messorte/steckbrief.php?ort=WALS; 51.496540, 6.748363
# - Essen-Vogelheim #https://www.lanuv.nrw.de/luqs/messorte/steckbrief.php?ort=EVOG; 51.496540, 6.980991
stations = Stations()
stations = stations.nearby(51.525955, 6.976880)
station = stations.fetch(1)
# Get daily data
data = Hourly(station, start, end)
data = data.fetch()
# Plot line chart including temperature, precipitation, dew point
# documentation: https://dev.meteostat.net/python/hourly.html#data-structure
data.plot(y=['temp', 'prcp', 'dwpt'])
plt.show()
print("Script completed!")
```

```python
start = datetime(2020, 1, 2)
end = datetime(2023, 10, 1)
stations = Stations()
stations = stations.nearby(51.496540, 6.748363)
station = stations.fetch(1)
dData = Hourly(station, start, end)
dData = dData.fetch()
dData.reset_index(inplace=True)
```
