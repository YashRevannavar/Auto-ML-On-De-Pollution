# Error Analysis

## Imports

```python
# %reset -f
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import plotly.express  as ex
sns.set()
rcParams['figure.figsize'] = 10,10
```

## Data Pre-processing

```python
df = pd.read_csv("../Data/cleanned/zusammen.csv", parse_dates=["date"])
# df['month'] = df['date'].dt.month
df['week'] = df['date'].dt.week
df.dropna(inplace=True)
df4 = pd.read_csv('../models/ml04/ml04Predictions.csv')
df41 = pd.read_csv('../models/ml04-1/ml04-1Predictions.csv')
```

```python
df0 = df[['date','week']]
df1 = df0[23005:]
df1.reset_index(drop=True, inplace=True)
df4.rename(columns={'prediction':'prediction4',
                    'Real':'real4',
                    'Error':'error4'}, inplace=True)
df41.rename(columns={'prediction':'prediction41',
                    'Real':'real41',
                    'Error':'error41'}, inplace=True)
df = pd.concat([df1, df4, df41], axis=1)
df = df.drop(index = 730, axis=1)
```

## Observations

```python
error04 = df.query('error4 < 0.1 and error4 > -0.1').count()[0]
error041 = df.query('error41 < 0.1 and error41 > -0.1').count()[0]
print(f'Errors between -0.1 and 0.1 with ML04: {error04}')
print(f'Errors between -0.1 and 0.1 with ML04.1: {error041}')
```

```python
ex.box(df,
       x='week',
       y='error4',
       title='Error Analysis',
       hover_name='error4',
       hover_data=['error4'],
       )
```

```python
ex.box(df,
       x='week',
       y='error41',
       title='Error Analysis',
       hover_name='error41',
       hover_data=['error41'],
       )
```

```python
ex.histogram(df, 
            x="error4", 
            nbins=100,
            title="Error from model 04")
```
