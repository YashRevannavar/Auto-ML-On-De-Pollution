# Auto-ML 04 with pycaret

```python
# %reset -f
import pandas as pd
from pycaret.regression import *
```

```python
df = pd.read_csv("../Data/cleanned/zusammen.csv", parse_dates=["date"])
df.drop(columns=["rohwert"], inplace=True)
df.dropna(inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
df
```

```python
s = setup(df, target = 'wert', session_id = 123)
```

```python
best = s.compare_models()
```

```python
print(best)
```

```python
s.save_model(best, '../models/aml-pycaret-03/aml-pycaret-03')
```

```python
best = load_model('../models/aml-pycaret-03/aml-pycaret-03')
print(best)
```

```python
predict = sb.predict_model(best)
```
